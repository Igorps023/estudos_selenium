```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd


from selenium.webdriver import Firefox
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep

from selenium.webdriver.support.ui import Select

url = 'https://www.ssp.sp.gov.br/estatistica/dados-mensais'
driver = Firefox()
driver.get(url)
sleep(1.2)


# Wait for the dynamic content to load (you may need to adjust the wait time)
driver.implicitly_wait(5)

# Get the updated HTML content
html_content = driver.page_source

# Close the browser
####driver.quit()

# Parse the updated HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

columns_xpath = "//div[@id='collapse0']//table[@class='table table-striped table-hover']/thead//th"
column_elements = driver.find_elements(By.XPATH, columns_xpath)

# Extracting text from each th element
column_headers = [column.text for column in column_elements]

rows_xpath = "//div[@id='collapse0']//table[@class='table table-striped table-hover']/tbody//tr"
rows_elements = driver.find_elements(By.XPATH, rows_xpath)

rows_data = []

for row in rows_elements:
    row_data = [cell.text for cell in row.find_elements(By.TAG_NAME, "th") + row.find_elements(By.TAG_NAME, "td")]
    rows_data.append(row_data)
```

```python
df = pd.DataFrame(rows_data, columns=column_headers)
df
```
