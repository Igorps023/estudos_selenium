```python
from selenium.webdriver import Firefox
from selenium.webdriver.common.by import By
from time import sleep
from urllib.parse import urlparse
from pprint import pprint

def acesso_tag_inicial(tag):
    elemento_inicial = browser.find_element(By.TAG_NAME, tag) #1 tag
    return elemento_inicial

def acesso_tag_composta(tag_inicial, item_procura):
    tag_composta = tag_inicial.find_elements(By.TAG_NAME, item_procura) #lista
    return tag_composta


def acesso_tag_especifica(tag_inicial, item_procura):
    tag_especifica = tag_inicial.find_element(By.TAG_NAME, item_procura) #1 elemento
    return tag_especifica

def filtro_atributo(tag_inicial, atributo, valor_procura_atributo):
    result = None
    for item in tag_inicial:
        attribute_value = item.get_attribute(atributo) #1 atributo
        if attribute_value and valor_procura_atributo in attribute_value:
            result = item
            break
    return result


url = 'https://curso-python-selenium.netlify.app/exercicio_03.html'
browser = Firefox()
browser.get(url)
sleep(1.2)

"""
PARTE 1
"""
elemento = acesso_tag_inicial('main')
ancora = acesso_tag_especifica(elemento, 'a')
ancora.click()

"""
PARTE 2
"""
elemento = acesso_tag_inicial('main')
ancora = acesso_tag_composta(elemento, 'a')
resultado = filtro_atributo(ancora, 'attr', 'errado')
resultado.click()

"""
PARTE 3
"""
elemento = acesso_tag_inicial('main')
ancora = acesso_tag_composta(elemento, 'a')
resultado = filtro_atributo(ancora, 'attr', 'certo')
resultado.click()

"""
PARTE 4
"""
url_path_atual = urlparse(browser.current_url).path
url_path_atual = url_path_atual.lstrip('/') #retira a '/' da string

elemento = acesso_tag_inicial('main')
ancora = acesso_tag_composta(elemento, 'a')
resultado = filtro_atributo(ancora, 'text', url_path_atual)
resultado.click()

"""
PARTE 4
"""
browser.refresh()
```
